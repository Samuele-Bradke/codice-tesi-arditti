## Analisis of time use

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tempi = pd.read_csv("block_timing_log.csv",index_col=0)
```

```python
columns = tempi.columns 
plt.stackplot(tempi.index, [tempi[col] for col in columns], labels=columns)
plt.title("Stacked Area Chart of Block Processing Times")
plt.xlabel("Block Number")
plt.ylabel("Time (s)")
plt.legend(loc='upper left')
plt.show()
```

```python
averages = tempi.mean()
plt.figure(figsize=(8, 8))
plt.pie(averages, labels=averages.index, autopct="%1.1f%%", startangle=140)
plt.title("Average Time Distribution Across Steps")
plt.show()
```
