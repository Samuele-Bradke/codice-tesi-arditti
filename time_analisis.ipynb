## Analisis of time use

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
tempi = pd.read_csv("block_timing_log.csv",index_col=0)
```

```python
columns = tempi.select_dtypes(include='number').columns 
plt.stackplot(tempi.index, [tempi[col] for col in columns], labels=columns)
plt.yscale('log')
plt.title("Stacked Area Chart of Block Processing Times (Log Scale)")
plt.xlabel("Block Number")
plt.ylabel("Time (s, log scale)")
plt.legend(loc='upper left')
plt.show()
```
